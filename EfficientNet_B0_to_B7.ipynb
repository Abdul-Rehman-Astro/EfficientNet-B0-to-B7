# 1: Image Classification Using EfficientNet B0 with Imagenet pretrained Model

```python
# To use EfficientNetB0 for classifying 1000 classes of images from imagenet, we can use Imagenet weights.

from tensorflow.keras.applications import EfficientNetB0
model = EfficientNetB0(weights='imagenet')  #trained on 1000 classes of Imagenet dataset
```

```python
# model.summary()
```

```python
import cv2
import numpy as np
from matplotlib.pyplot import imread
from matplotlib.pyplot import imshow
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.imagenet_utils import decode_predictions
from tensorflow.keras.applications.imagenet_utils import preprocess_input


img_path = '/content/drive/MyDrive/Spark IIT Roorkee/B0 to B7/cat.jpg'

img = cv2.imread(img_path)
img = cv2.resize(img, (224, 224)) # Notice the image size for B0

x = np.expand_dims(img, axis=0)
x = preprocess_input(x)

print('Input image shape:', x.shape)

my_image = imread(img_path)
imshow(my_image)
```

We will find probability of above image being in each of these 1000 classes

```python
predsb0=model.predict(x)
# print("predicted class: ", predsb0 )    # probabilities for being in each of the 1000 classes
```

```python
decode_predictions(predsb0,top=3)  #  # probabilities for being in top of the 3 classes
```

Notice that top 3 probabilities are cat.






1.   Egyptian_cat , 82.8 %
2.   tabby, 5.5 %
3.   tiger_cat , 2.8 %
